```python
import pandas as pd
import numpy as np
from scipy import stats

model = pd.read_csv('model.csv')
player_dict = {}
for i, player in enumerate(model['Player']):
    player_dict[i] = player
player_dict
```

```python
print(model.columns)
metrics = ['Cmp%_zscore', 'TD%_zscore', 'Int%_zscore', 'Y/G_zscore', 'Sk%_zscore', 'ANY/A_zscore']
```

```python
from sklearn.neighbors import NearestNeighbors
X = np.array(model.loc[:, metrics])
nbrs = NearestNeighbors().fit(X)
distance, indices = nbrs.kneighbors([X[78]], n_neighbors=11)
indices = list(indices.flatten())
for i in indices:
    print(player_dict[i])
```
