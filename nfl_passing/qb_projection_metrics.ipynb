```python
import numpy as np
import pandas as pd
from sklearn import metrics
```

```python
# Read in data
df = pd.read_csv('regression_data.csv', index_col='Player')
df = df.drop(columns='Unnamed: 0')
df.query('Year == 2019')
```

```python
y_true = np.array(df['ANY/A_zscore_actual'])
y_predict = np.array(df['ANY/A_zscore_predict'])
mse = metrics.mean_squared_error(y_true, y_predict)
rmse = round(np.sqrt(mse), 2)
rmse
```
