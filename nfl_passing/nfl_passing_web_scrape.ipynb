```python
import pandas as pd

dfs = pd.read_html('https://www.pro-football-reference.com/years/2019/passing.htm')
passing = dfs[0]
passing.shape
```

```python
passing.columns
```

```python
passing_qbs = passing.query('Pos in ["QB", "qb"]')
passing_qbs
```

```python
passing_2019 = passing_qbs.drop(labels=['Rk', 'Pos', 'QBR', '4QC', 'GWD'], axis=1)
passing_2019.columns
```

```python
clean = passing_2019.rename(columns={'Yds.1': 'SkYds'})
clean.columns
```

```python
clean['Year'] = 2019
clean
```

```python
years = [year for year in range(1982, 2020)]
passing_stats = []
for year in years:
    url = 'https://www.pro-football-reference.com/years/' + str(year) + '/passing.htm'
    dfs = pd.read_html(url)
    passing = dfs[0]
    passing_qbs = passing.query("Pos in ['QB', 'qb']")
    drop_columns = passing_qbs.drop(labels=['Rk', 'Pos'], axis=1)
    rename_columns = drop_columns.rename(columns={'Yds.1': 'SkYds'})
    rename_columns['Year'] = year
    passing_stats.append(rename_columns)

all_passing = pd.concat(passing_stats, sort=False)
all_passing.shape
```

```python
all_passing.query('Tm == "DAL"')
```

```python
all_passing.to_csv('nfl_passing_1982_2019.csv')
```
