```python
import pandas as pd

dfs = pd.read_html('https://www.pro-football-reference.com/years/NFL/passing.htm')
league_passing = dfs[0]
league_passing.columns
```

```python
league_passing.columns = league_passing.columns.to_flat_index()
league_passing.columns
```

```python
league_passing.columns = [col[1] for col in league_passing.columns]
league_passing.head()
```

```python
league_passing.columns
```

```python
league_passing = league_passing.drop(columns=['Rk', 'Tms'])
league_passing.head()
```

```python
league_passing_num = league_passing.apply(pd.to_numeric, errors='coerce')
league_passing_num.head()
```

```python
passing_1982 = league_passing_num.query('Year >= 1982')
passing_1982.loc[:, 'Year'] = passing_1982.loc[:, 'Year'].astype('int16')
passing_1982
```

```python
passing_1982.to_csv('average_passing_stats_1982_2019.csv', index=False)
```
